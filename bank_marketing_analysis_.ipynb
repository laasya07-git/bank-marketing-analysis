```python
# Load the CSV file
import pandas as pd

df = pd.read_csv(r"C:\Users\DELL\Downloads\bankmarketing.csv")
df.head()
```

```python
#getting information about the dataset
df.info()
```

```python
#checking missing values
df.isnull().sum()
```

```python
#describe the data
df.describe(include='all')
```

```python
#class distribution
df['y'].value_counts(normalize=True) * 100
```

```python
#visualizations
import seaborn as sns
import matplotlib.pyplot as plt

# Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Job vs Subscription
plt.figure(figsize=(10, 5))
sns.countplot(x='job', hue='y', data=df)
plt.xticks(rotation=45)
plt.title('Job vs Term Deposit Subscription')
plt.show()
```

```python
# Marital Status vs Subscription
plt.figure(figsize=(6, 4))
sns.countplot(x='marital', hue='y', data=df)
plt.title('Marital Status vs Subscription')
plt.show()
```

```python
import numpy as np

#numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Plotting correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='campaign', hue='y', data=df, alpha=0.7)
plt.title("Age vs Campaign Contacts Colored by Subscription")
plt.show()
```

```python
#boxplot 
plt.figure(figsize=(12, 6))
sns.boxplot(x='job', y='age', data=df)
plt.xticks(rotation=45)
plt.title("Boxplot of Age by Job Type")
plt.show()
```

```python
#countplot
plt.figure(figsize=(8, 4))
sns.countplot(x='education', data=df, order=df['education'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Count of Education Levels")
plt.show()
```
